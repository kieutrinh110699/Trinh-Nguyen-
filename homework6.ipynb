```python
from sklearn.linear_model import LogisticRegression
import statsmodels.api as sm
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
df = pd.read_csv('framingham.csv')
```

```python
#1.
df.describe()
df['education']=df['education'].fillna(df['education'].median())
df['cigsPerDay']=df['cigsPerDay'].fillna(df['cigsPerDay'].median())
df['BPMeds']=df['BPMeds'].fillna(df['BPMeds'].median())
df['totChol']=df['totChol'].fillna(df['totChol'].median())
df['BMI']=df['BMI'].fillna(df['BMI'].median())
df['heartRate']=df['heartRate'].fillna(df['heartRate'].median())
df['glucose']=df['glucose'].fillna(df['glucose'].median())
```

```python
df.describe()
```

```python
#2
#male will increase the chance of being diabetes, high BMI, heart rate and high glucose.
#age will increase the chance of being diabetes, high BMI, heart rate and high glucose.
#education does not affect the chance of being diabetes, high BMI, heart rate and high glucose.
#smoke will increase the chance of being diabetes, high BMI, heart rate and high glucose.
#more ciggeretes will increase the chance of being diabetes, high BMI, heart rate and high glucose.
#being on medication will increase the chance of being diabetes, high BMI, heart rate and high glucose.
#Stroke will increase the chance of being diabetes, high BMI, heart rate and high glucose.
#High Blood Pressure will increase the chance of being diabetes, high BMI, heart rate and high glucose.
#Diabetes will increase the chance of being high BMI, heart rate and high glucose.
#totChol, sysBP diaBP will increase the chance of being diabetes, high BMI, heart rate and high glucose.
#BMI will increase the chance of being diabete, heart rate and high glucose
#hear rate will increase the chance of being diabetes, high BMI and high glucose
#high glucose will increase the chance of being diabetes, high BMI, heart rate.
```

```python
#3
X=df[['male','age','education','currentSmoker','cigsPerDay','BPMeds','prevalentStroke','prevalentHyp','diabetes','totChol','sysBP','diaBP','BMI','heartRate','glucose']]
X = sm.add_constant(X)
y=df['TenYearCHD']
```

```python
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i)
                          for i in range(len(X.columns))]
print(vif_data)
```

```python
#4
model = sm.Logit(y, X).fit()
y_pred = model.predict(X)
resid = y - y_pred
print(model.summary())
```
