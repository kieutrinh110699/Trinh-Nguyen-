```python
stop_words = stopwords.words('english')
stop_words
```

```python
items1 = blob1.word_counts.items()
items2 = blob2.word_counts.items()
items1 = [item for item in items1 if item[0] not in stop_words]
items2 = [item for item in items2 if item[0] not in stop_words]
```

```python
sorted_items1 = sorted(items1, key=itemgetter(1), reverse=True)
top20 = sorted_items1[1:21]
top20
```

```python
sorted_items2 = sorted(items2, key=itemgetter(1), reverse=True)
top20 = sorted_items2[1:21]
top20
```

```python
df = pd.DataFrame(top20, columns=['word', 'count'])
axes = df.plot.bar(x='word', y='count', legend=False)
plt.gcf().tight_layout()
```
