```python
from sklearn.linear_model import LogisticRegression
import statsmodels.api as sm
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
#1. 
df = pd.read_csv('club.csv')
```

```python
df.head(6)
```

```python
#1b
df['credit_card']=pd.get_dummies(df['purpose'])['credit_card']
```

```python
pd.get_dummies(df['purpose'])
```

```python
df['revol_util']=df['revol_util'].fillna(df['revol_util'].median())
```

```python
df.describe()
```

```python
#1c
X = df[['credit_policy','int_rate','log_annual_income','dti','fico','days_with_cr_line',
        'revol_bal','revol_util','inq_last_6mths','delinq_2yrs','pub_rec','credit_card']]
X = sm.add_constant(X)
y = df[['not_fully_paid']]
```

```python
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i)
                          for i in range(len(X.columns))]
print(vif_data)
```

```python
#1D
X = df[['credit_policy','int_rate','log_annual_income','dti','fico','days_with_cr_line',
        'revol_bal','revol_util','inq_last_6mths','delinq_2yrs','pub_rec','credit_card']]
X = sm.add_constant(X)
y = df[['not_fully_paid']]
model = sm.Logit(y, X).fit()
print(model.summary())
```
